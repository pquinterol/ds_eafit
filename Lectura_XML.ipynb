En este codigo se busca encontrar que objects ID tienen xml de los cuales se pueda minar la data disponibles. Se le debe de entregar un archivo csv que tenga una columna llamada "ein".

```python
def txtCreator():
  #Busqueda del archivo de ein en S3 que vamos a usar.
  f = open('/content/einList.csv', 'r')

  data_pandas = pd.read_csv(f)
  #Definicion de la lista de index que se tiene en s3 para hacer el filtrado de objects id.
  files = os.listdir("csv")
  numb = 6
        
  if(not os.path.isdir("irs_files")):
    os.mkdir("irs_files")
  else:
    shutil.rmtree("irs_files")
    mkdir("irs_files")
  #Ciclo que recorre todos los archivos.
  for file in sorted(files):
      lista = []
      pathh = f'csv/{file}'
      print(pathh)
      data_id = pd.read_csv(pathh)
      #Query para sacar los object id de aquellos ein que se encuentran en el archivo dado anteriormente.
      data = data_id.query('EIN in @data_pandas.ein').OBJECT_ID

      #Este numero es solo para sacar la notificacion de cuantos ids validos hay por año. 
      numbe = "201" + str(numb)
      if numb >= 10:
          numbe = "202" + str(numb%10)
      os.mkdir(f"irs_files/_{numbe}")
      patth = f"irs_files/_{numbe}/{numbe}.txt"
      for element in data:
          lista.append(str(element))
      print(f"Se encontraron :{len(lista)} ids buenos para trabajar en el año: {numbe}")
      with open(patth, 'w') as f:
          f.write(",".join(lista))
      numb+=1
```

```python
from datetime import datetime

def separator_main(anio):
    if anio != 500:
        pathh = f"irs_files/_{anio}/{anio}.txt"
        with open(pathh, "r") as f:
            lista_de_index_numeros = f.read().split(',')
        obtain_valid_object_id([0,len(lista_de_index_numeros)-1,0],anio)
        separator_valid(anio)
    
def separator_valid(anio):
    xml_runner = XMLRunner()
    valid_list = []
    pathh = f'irs_files/_{anio}/funcional.txt'
    with open(pathh, 'r') as f:
        list_ids = f.read().split(',')
    iteracion= 0
    for i in list_ids:
        if iteracion % 1000 == 0:
            print(f" === {iteracion}/{len(list_ids)}", end=" === ")
        parsed_filing = xml_runner.run_filing(i)
        if parsed_filing.get_version() in ALLOWED_VERSIONSTRINGS:
            valid_list.append(i)
        iteracion+=1
    pathh =f'irs_files/_{anio}/funcional_valid.txt'
    with open(pathh, 'w') as f:
        f.write(",".join(valid_list))



def obtain_valid_object_id(list_parameters,anio):

    lista_de_index_numeros = []
    pathh = f"irs_files/_{anio}/{anio}.txt"
    with open(pathh, "r") as f:
        lista_de_index_numeros = f.read().split(',')
    contador_malos = 0

    contador_buenos = 0

    iteracion = 0

    functional_id = []

    nonfunctional_id = []

    xml_runner = XMLRunner()
    start_time = datetime.now()
    for i in lista_de_index_numeros[list_parameters[0]:list_parameters[1]]:

        if iteracion % 1000 == 0:
            print(f"{iteracion}/{list_parameters[1]}", end=" === ")
        try:
            parsed_filing = xml_runner.run_filing(i)
            result = parsed_filing.get_result()  # An array of parsed schedules
            schedule_list = parsed_filing.list_schedules()  # an array of form names
            contador_buenos += 1
            functional_id.append(str(i))
        except:
            contador_malos += 1
            nonfunctional_id.append(str(i))
        iteracion += 1
    end_time = datetime.now()

    print(
        f"\n ==== Ejecución: {list_parameters[2]} | Rango: [{list_parameters[0]}, {list_parameters[1]}) | Tiempo: {end_time - start_time} ===="
    )

    file_s_time = datetime.now()
    pathh= f'irs_files/_{anio}/nonFunctional.txt'
    with open(pathh, 'w') as f:
        f.write(",".join(nonfunctional_id))
    pathh = f'irs_files/_{anio}/funcional.txt'
    with open(pathh, 'w') as f:
        f.write(",".join(functional_id))
    pathh=f'irs_files/_{anio}/datos_de_perdida_y_total.txt'
    with open(pathh, 'w') as f:
        f.writelines("Se encontrarone en total :" +
                     str(contador_buenos) + " ID en buen estado")
        f.writelines("\nSe encontraron en total :" +
                     str(contador_malos) + " ID en mal estado")
    file_e_time = datetime.now()
    print(
        f"""
*** 
Ejecución: {list_parameters[2]}
Tiempo de escritura: {file_e_time - file_s_time} 
Rango [{list_parameters[0]}, {list_parameters[1]})
***
        """
    )
```

```python
import json
import csv


from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

from webdriver_manager.chrome import ChromeDriverManager


def variable_name_and_description_to_csv(parameters_list):
    """
        Mediande identificadores hallar todos los nombres y descripciones, para posteriormente guardar esto en un único csv
        Parametros
            url: (str) con la url de la cuál se van a obtener los datos
            file_name: (str) nombre con el que se va a guardar el csv
            show_browser: (bool) abrir o no chrome al momento de realizar la ejecución
    """
    url = parameters_list[0]
    file_name = parameters_list[1]
    show_browser = parameters_list[2]
    delimiter = parameters_list[3]

    chrome_options = Options()

    if show_browser:
        chrome_options.add_argument("start-maximized")
    else:
        chrome_options.add_argument('headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    wd = webdriver.Chrome('chromedriver',options=chrome_options)

    wd.get(url)

    # Sacar todos los Variable Name:
    variable_all_data_a_list = wd.find_elements(
        By.XPATH, value='//p[@class="h3 indent1"]/a')

    # Sacar todos los Description
    variable_all_data_p_list = wd.find_elements(
        By.XPATH, value='//p[@class="lead indent2"]')

    # Sacar solo el texto
    variable_name_list = []
    variable_description_list = []

    # Obtener los textos de los anchor únicamente
    for name in variable_all_data_a_list:
        variable_name_list.append(name.text)

    # Obtener el texto de la descripción únicamente
    for description in variable_all_data_p_list:
        variable_description_list.append(description.text[description.text.find(
            "Description:")+13:description.text.find("\nType")])

    # Convertir ambas listas a una lista
    variable_name_and_description = list(
        zip(variable_name_list, variable_description_list))
    


    # Path y nombre del archivo
    path_name_and_description = f"irs_file/IRS990/csv/variable_name_and_description-{file_name}.csv"

  
    
    # Path solo nombre del archivo
    path_name = f"irs_file/IRS990/csv/variable_name-{file_name}.csv"

    # Tamaño
    print(f"{file_name} tamaño {len(variable_name_list)}")

    # Guardar la lista de listas con nombre,descripción en csv
    with open(path_name_and_description, 'w') as f:
        writer = csv.writer(f, delimiter=delimiter, lineterminator="\n")
        writer.writerows(variable_name_and_description)

    # Guardar la lista de nombres
    with open(path_name, 'w') as f:
        f.write(json.dumps(variable_name_list))

    # Cerrar el driver para guardar recursos
    wd.close()

# Creacion de directorio
if(not os.path.isdir("irs_file/")):
    os.mkdir("irs_file/")
else:
    shutil.rmtree("irs_file")
    os.mkdir("irs_file/")
    
if(not os.path.isdir("irs_file/IRS990")):
    os.mkdir("irs_file/IRS990")

if(not os.path.isdir("irs_file/IRS990/csv")):
    os.mkdir("irs_file/IRS990/csv")
variable_name_and_description_to_csv(
    ['http://www.irsx.info/metadata/parts/part_0.html', 'part_0', False, '|']
)



variable_name_and_description_to_csv(
    ['http://www.irsx.info/metadata/parts/part_i.html', 'part_i', False, '|']
)


variable_name_and_description_to_csv(
    ['http://www.irsx.info/metadata/parts/part_iii.html', 'part_iii', False, '|']
)



variable_name_and_description_to_csv(
    ['http://www.irsx.info/metadata/parts/part_iv.html', 'part_iv', False, '|']
)

variable_name_and_description_to_csv(
    ['http://www.irsx.info/metadata/parts/part_v.html', 'part_v', False, '|']
)

variable_name_and_description_to_csv(
    ['http://www.irsx.info/metadata/parts/part_vi.html', 'part_vi', False, '|']
)


variable_name_and_description_to_csv(
    ['http://www.irsx.info/metadata/parts/part_vii.html', 'part_vii', False, '|']
)


variable_name_and_description_to_csv(
    ['http://www.irsx.info/metadata/parts/part_viii.html', 'part_viii', False, '|']
)


variable_name_and_description_to_csv(
    ['http://www.irsx.info/metadata/parts/part_ix.html', 'part_ix', False, '|']
)



variable_name_and_description_to_csv(
    ['http://www.irsx.info/metadata/parts/part_x.html', 'part_x', False, '|']
)

variable_name_and_description_to_csv(
    ['http://www.irsx.info/metadata/parts/part_xi.html', 'part_xi', False, '|']
)

variable_name_and_description_to_csv(
    ['http://www.irsx.info/metadata/parts/part_xii.html', 'part_xii', False, '|']
)



```

```python
def matrix_creation_sked_parts(object_id, title, anio):
    xml_runner = XMLRunner()
    parsed_filing = xml_runner.run_filing(object_id)
    list_schedules = parsed_filing.list_schedules()
    skedc = 'part_0'
    groupb = 'schedule_parts'
    data_temp = ''
    counter_len = 0
    array_prt = []  # Fila con los headers ['object_id','Schedule','...']
    # Sacar todo lo que tiene part de array_prt en una lista

    array_prt.count("NOMBRE-del-GRUPO")

    datos = []  # Lista con los datos individuales que se van a llenar
    datos.append(object_id)
    datos.append(parsed_filing.ein)
    part_0_list_len, part_i_list_len = lens(anio)[0], lens(anio)[1]
    part_iii_list_len, part_iv_list_len, part_v_list_len = lens(anio)[2], lens(anio)[
        3], lens(anio)[4]
    part_vi_list_len, part_vii_list_len, part_viii_list_len, part_ix_list_len = lens(anio)[
        5], lens(anio)[6], lens(anio)[7], lens(anio)[8]
    part_x_list_len, part_xi_list_len, part_xii_list_len = lens(anio)[9], lens(anio)[
        10], lens(anio)[11]

    # Longitudes de grupos!

    array_prt = lens(anio)[12]

    # Creacion listas de uso

    matriz_solo_datos = []
    array_total = array_prt

    # Creacion variables numeros:
    numero1 = part_0_list_len
    numero2 = numero1+part_i_list_len
    numero3 = numero2+part_iii_list_len
    numero4 = numero3+part_iv_list_len
    numero5 = numero4+part_v_list_len
    numero6 = numero5+part_vi_list_len
    numero7 = numero6+part_vii_list_len
    numero8 = numero7+part_viii_list_len
    numero9 = numero8+part_ix_list_len
    numero10 = numero9+part_x_list_len
    numero11 = numero10+part_xi_list_len
    numero12 = numero11+part_xii_list_len


    if "IRS990" in list_schedules:
        parsed_filing = xml_runner.run_sked(
            object_id, 'IRS990'
        )

        result = parsed_filing.get_result()

        sked_temp = ''
        int_skedc = 0
        for data in array_total:

            data_list = []

            if counter_len >= numero1 and counter_len < numero2:
                skedc = 'part_i'
            elif counter_len >= numero2 and counter_len < numero3:
                skedc = 'part_iii'
            elif counter_len >= numero3 and counter_len < numero4:
                skedc = 'part_iv'
            elif counter_len >= numero4 and counter_len < numero5:
                skedc = 'part_v'
            elif counter_len >= numero5 and counter_len < numero6:
                skedc = 'part_vi'
            elif counter_len >= numero6 and counter_len < numero7:
                skedc = 'part_vii'
            elif counter_len >= numero7 and counter_len < numero8:
                skedc = 'part_viii'
            elif counter_len >= numero8 and counter_len < numero9:
                skedc = 'part_ix'
            elif counter_len >= numero9 and counter_len < numero10:
                skedc = 'part_x'
            elif counter_len >= numero10 and counter_len < numero11:
                skedc = 'part_xi'
            elif counter_len >= numero11 and counter_len < numero12:
                skedc = 'part_xii'

            try:
                if groupb == 'schedule_parts':
                    datos.append(result[0][groupb][skedc][data])

            except:
                datos.append('-')

            counter_len += 1

        matriz_solo_datos.append(datos)
    if title:

        return ['object_id', 'ein']+array_total
    return matriz_solo_datos
```

# Creacion de csv partiendo de la extraccion de los datos.

```python
def crear_csv_IRS990(anio):
    valid_path_IRS990 = f"irs_files/_{anio}/funcional_valid.txt"
    valid_IRS990_object_id = []

    with open(valid_path_IRS990, "r") as f:
        valid_IRS990_object_id = f.read().split(',')


    title = []
    all_list = lens(anio)[12]
    matriz = []
    counter_text = 0
    iteracion = 0
    print("Iniciando proceso de creacion de excel para las partes.")
    for i in valid_IRS990_object_id:

        print(f" === {iteracion}/{len(valid_IRS990_object_id)}", end=" === ")

        if counter_text == 0:
            title = matrix_creation_sked_parts(
                i,  True, anio)
        matriz.append(matrix_creation_sked_parts(
            i,  False, anio))
        iteracion += 1

    pathh = f"irs_results/{anio}/IRS990/Parts/IRS990.csv"
    with open(pathh, 'a') as f:
        writer = csv.writer(f, delimiter=",", lineterminator="\n")
        writer.writerow(title)
        for roww in matriz:
            writer.writerows(roww)
```
