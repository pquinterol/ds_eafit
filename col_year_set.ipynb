```
from pyspark.sql import DataFrame
import pyspark.sql.functions as F
from pyspark.sql.types import *
```

```
def get_data(path: str, file_fomat: str) -> DataFrame:
    df = (spark
          .read
          .format(file_fomat)
          .option("header", "true")
          .option("inferSchema", "true")
          .load(path)
         )
    
    return df
```

```
def set_year(df: DataFrame, year: str) -> DataFrame:
    data = df.withColumn("year", F.lit(year))
    return data
```

```
def write_data(df: DataFrame, path: str, file_format: str = "csv"):
    (df
     .coalesce(1)
     .write
     .format(file_format)
     .option("header", "true")
     .save(path)
    )
```

```
for year in range (2015, 2021):
    
    data_read_path = f"s3://datalake-ds-project/data/raw/{year}/IRS990/Parts/IRS990.csv"
    data_write_path = f"dbfs:/FileStore/tables/tmp/EAFIT/ds/uncleaned/IRS990_{year}"
    
    data = get_data(data_read_path, "csv")
    year_tagged_data = set_year(data, year)
    write_data(year_tagged_data, data_write_path, "csv")
```
